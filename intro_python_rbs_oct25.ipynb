```python
exam_score = 95.9997
```

```python
h_w_ratio = 0.8
```

```python
# String Output - F-string (variable-interpolation capable)
```

```python
print(f'My name is {my_name} and I am {my_age} years old.')
```

```python
print(f'My name is {my_name} and I am {my_age} years old. I scored {exam_score: .2f} in the exam. My h_w_ratio is{h_w_ratio: .0%}')
```

```python
print(f'My name is {my_name} and I am {my_age} years old. I scored {exam_score: .4f} in the exam. My h_w_ratio is{h_w_ratio: .2%}')
```
